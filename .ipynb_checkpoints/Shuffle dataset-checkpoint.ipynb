```python
%matplotlib inline

import pandas as pd
import numpy as np
import random
from pandas import ExcelWriter
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("h1b.csv")
dataset = dataset.sample(frac=0.02).reset_index(drop=True)
dataset.head()
```

```python
treina,teste = train_test_split(dataset, test_size=0.2)
```

```python
treina = treina.applymap(lambda x: x.encode('unicode_escape').decode('utf-8') if isinstance(x, str) else x)
print(len(treina))
treina.head()
```

```python
teste = teste.applymap(lambda x: x.encode('unicode_escape').decode('utf-8') if isinstance(x, str) else x)
print(len(teste))
teste.head()
```

```python
writer = pd.ExcelWriter('Dataframe.xlsx')
treina.to_excel(writer,'treinamento')
teste.to_excel(writer,'teste')
writer.save()
```
